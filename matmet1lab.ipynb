```python
import io
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel(io.BytesIO(uploaded['Book2 (1).xlsx']))
```

```python
df.dtypes
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

#lin reg

```python
model=LinearRegression()
```

```python
x=pd.DataFrame(df['x'])
y=pd.DataFrame(df['y'])
```

```python
model.fit(x,y)
```

```python
print('liner reg',model.score(x,y))
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),c='r')
```

#poly reg

```python
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)
```

```python
model2 = LinearRegression()
model2.fit(x_poly, y)
y_poly_pred = model2.predict(x_poly)
```

```python
print('poly reg',model2.score(x_poly,y))
```

```python
plt.scatter(x,y)
plt.plot(x,model2.predict(x_poly),c='r')
```

#log reg

```python
ln_x = np.log(x)
```

```python
model3 = LinearRegression()
model3.fit(ln_x, y)
y_log_pred = model3.predict(ln_x)
```

```python
model3.score(ln_x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model3.predict(ln_x),c='r')
```

#exp reg

```python
y_log=np.log(y)
model4=LinearRegression()
model4.fit(x,y_log)
print(model4.score(x,y_log))
```

```python
plt.scatter(x,y)
plt.plot(x,np.exp(model4.predict(x)),c='r')
```
